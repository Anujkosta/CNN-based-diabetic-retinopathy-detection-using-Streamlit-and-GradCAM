# 🧠 Grad-CAM Explainable AI (XAI) for Diabetic Retinopathy

---

## 📖 What is Grad-CAM? (Simple Explanation for Viva)

**Grad-CAM (Gradient-weighted Class Activation Mapping)** is a technique that
helps us **understand what part of an image a neural network is looking at**
when it makes a prediction.

### 🔍 How it works — in plain English:

Imagine the CNN as a detective solving a mystery (classifying your image).
After the detective gives its verdict (e.g., "Moderate DR"), Grad-CAM
asks: **"Which clues (pixels/regions) most influenced your decision?"**

**Step-by-step:**
1. **Forward pass** — The image goes through the network; we get a prediction.
2. **Pick the target class** — Usually the predicted class (or any class you want to explain).
3. **Compute gradients** — We calculate how much each feature map in the *last
   convolutional layer* contributes to that class score. These are the "gradients".
4. **Global Average Pooling of gradients** — We average the gradients across
   spatial dimensions to get a single importance score (weight) per feature map.
5. **Weighted sum of feature maps** — Multiply each feature map by its weight
   and sum them all up → this gives a raw heatmap.
6. **ReLU + Normalize** — Apply ReLU (keep only positive activations = important
   regions) and normalize to [0, 1].
7. **Overlay** — Resize and overlay the heatmap on the original image using a
   color map (jet/rainbow) so hot regions (red) = most important.

### 🎯 Why it's useful for DR:
- Confirms the model is looking at **actual retinal lesions** (microaneurysms,
  hemorrhages, exudates) rather than irrelevant background.
- Builds **trust** in the model's predictions for clinical use.
- Helps identify model **failure modes** when it focuses on wrong areas.

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 1 — Imports
# ═══════════════════════════════════════════════════════════════════════════════

import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import os

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import load_model

print(f"TensorFlow version: {tf.__version__}")
print("All imports successful ✅")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 2 — Configuration (change paths here)
# ═══════════════════════════════════════════════════════════════════════════════

# ── Paths ──────────────────────────────────────────────────────────────────────
MODEL_PATH   = 'best_model.keras'         # Path to your saved model
TEST_IMG_DIR = 'F:/DR_Project/data/train_images/'  # Folder with test images

# ── Model settings ─────────────────────────────────────────────────────────────
IMG_SIZE     = (224, 224)                 # Must match your training size
NUM_CLASSES  = 5

# ── Class names in the EXACT order used during training ────────────────────────
# (alphabetical order — matches ImageDataGenerator default)
CLASS_NAMES  = ["Mild", "Moderate", "No_DR", "Proliferate_DR", "Severe"]

# ── Grad-CAM target layer name inside EfficientNetB0 ──────────────────────────
# 'top_conv' is the last conv layer before GlobalAveragePooling in EfficientNetB0
GRADCAM_LAYER = 'top_conv'

print("Configuration loaded ✅")
print(f"  Model path : {MODEL_PATH}")
print(f"  Image size : {IMG_SIZE}")
print(f"  Classes    : {CLASS_NAMES}")
print(f"  Grad-CAM layer: {GRADCAM_LAYER}")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 3 — Load Trained Model
# ═══════════════════════════════════════════════════════════════════════════════

def load_trained_model(model_path):
    """
    Loads a saved Keras model from disk.
    Works for both Sequential and Functional API models.

    Parameters:
        model_path (str): Path to the .keras or .h5 model file.

    Returns:
        model: Loaded Keras model.
    """
    # We pass compile=False so it loads even if custom loss is not registered
    model = load_model(model_path, compile=False)
    print(f"Model loaded successfully from: {model_path} ✅")
    print(f"Model type: {type(model).__name__}")
    print(f"Total layers: {len(model.layers)}")
    return model


# Load the model
model = load_trained_model(MODEL_PATH)
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 4 — Find the Correct Grad-CAM Layer
# ═══════════════════════════════════════════════════════════════════════════════

def find_gradcam_layer(model, target_layer_name=None):
    """
    Finds the best convolutional layer for Grad-CAM.

    Strategy:
    - If target_layer_name is given and exists → use it.
    - If model is Sequential → search inside sub-models.
    - Fallback: use the last Conv2D layer found anywhere in the model.

    Parameters:
        model: Keras model (Sequential or Functional).
        target_layer_name (str): Name of the desired layer.

    Returns:
        str: Name of the layer to use for Grad-CAM.
    """
    all_layer_names = [layer.name for layer in model.layers]

    # Case 1: Direct layer found in top-level model
    if target_layer_name and target_layer_name in all_layer_names:
        print(f"✅ Found target layer directly: '{target_layer_name}'")
        return target_layer_name

    # Case 2: Model contains a sub-model (e.g. Sequential wrapping EfficientNet)
    for layer in model.layers:
        if hasattr(layer, 'layers'):  # It's a sub-model
            sub_layer_names = [sl.name for sl in layer.layers]
            if target_layer_name and target_layer_name in sub_layer_names:
                print(f"✅ Found '{target_layer_name}' inside sub-model '{layer.name}'")
                return layer.name  # We return the sub-model's name for Grad-CAM extraction

    # Case 3: Fallback — find the last Conv2D in the entire model
    print(f"⚠️  Layer '{target_layer_name}' not found directly. Searching for last Conv2D...")
    last_conv = None
    for layer in model.layers:
        if isinstance(layer, (tf.keras.layers.Conv2D,)):
            last_conv = layer.name
        # Also check inside nested models
        if hasattr(layer, 'layers'):
            for sub_layer in layer.layers:
                if isinstance(sub_layer, (tf.keras.layers.Conv2D,)):
                    last_conv = layer.name  # Use sub-model name

    if last_conv:
        print(f"✅ Using fallback layer: '{last_conv}'")
        return last_conv

    raise ValueError("❌ Could not find any suitable Conv2D layer for Grad-CAM!")


def print_model_layers(model, max_layers=20):
    """Prints all layers and their names for inspection."""
    print("\n── Top-level layers ──────────────────────────")
    for i, layer in enumerate(model.layers):
        print(f"  [{i:3d}] {layer.name:<40} | {type(layer).__name__}")

    # If model wraps another model, also print that
    for layer in model.layers:
        if hasattr(layer, 'layers') and len(layer.layers) > 0:
            print(f"\n── Layers inside sub-model: '{layer.name}' (showing last {max_layers}) ─")
            for i, sub_layer in enumerate(layer.layers[-max_layers:]):
                print(f"  [{i:3d}] {sub_layer.name:<40} | {type(sub_layer).__name__}")


# Print layers so you can identify the correct one
print_model_layers(model)

# Find the Grad-CAM layer automatically
gradcam_layer_name = find_gradcam_layer(model, GRADCAM_LAYER)
print(f"\n🎯 Will use layer for Grad-CAM: '{gradcam_layer_name}'")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 5 — Image Preprocessing Function
# ═══════════════════════════════════════════════════════════════════════════════

def preprocess_image(image_path, img_size=(224, 224)):
    """
    Loads and preprocesses a retinal image for EfficientNetB0.

    EfficientNetB0 expects:
    - RGB image (not BGR)
    - Pixel values in range [0, 255] as float32
      (EfficientNet has its own internal normalization)

    Parameters:
        image_path (str): Full path to the image file.
        img_size (tuple): Target size, default (224, 224).

    Returns:
        img_array (np.ndarray): Preprocessed image with batch dim,
                                shape (1, H, W, 3), dtype float32.
        original_img (np.ndarray): Original RGB image (H, W, 3) uint8,
                                   for visualization.
    """
    # Step 1: Read the image with OpenCV (reads as BGR)
    img_bgr = cv2.imread(image_path)
    if img_bgr is None:
        raise FileNotFoundError(f"❌ Image not found at: {image_path}")

    # Step 2: Convert BGR → RGB (OpenCV loads as BGR by default)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # Step 3: Resize to the expected input size
    img_resized = cv2.resize(img_rgb, img_size)

    # Step 4: Keep a copy of the original uint8 image for visualization
    original_img = img_resized.copy()

    # Step 5: Convert to float32 — stay in [0, 255] range for EfficientNet
    img_float = img_resized.astype(np.float32)

    # Step 6: Add batch dimension → shape becomes (1, 224, 224, 3)
    img_array = np.expand_dims(img_float, axis=0)

    return img_array, original_img


print("preprocess_image() function defined ✅")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 6 — Grad-CAM Heatmap Generation Function
# ═══════════════════════════════════════════════════════════════════════════════

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    """
    Generates a Grad-CAM heatmap for a given image and model.

    How it works:
    1. Creates a sub-model that outputs BOTH the feature maps of the last
       conv layer AND the final predictions.
    2. Uses tf.GradientTape to record gradients of the predicted class score
       with respect to the last conv layer's output.
    3. Pools the gradients (Global Average Pooling) → importance weights.
    4. Multiplies feature maps by their weights and sums → raw heatmap.
    5. Applies ReLU (only keep positive = class-specific activations).
    6. Normalizes to [0, 1].

    Parameters:
        img_array (np.ndarray)    : Preprocessed image with batch dim (1,H,W,3).
        model (keras.Model)       : Trained Keras model.
        last_conv_layer_name (str): Name of the target convolutional layer.
        pred_index (int or None)  : Class index to explain. If None, uses
                                    the predicted (argmax) class.

    Returns:
        heatmap (np.ndarray): 2D heatmap normalized to [0, 1], shape (H', W').
        pred_index (int)    : The class index that was explained.
        pred_probs (np.ndarray): Full probability vector for all classes.
    """

    # ── Step 1: Handle both Sequential and Functional models ──────────────────
    # For Sequential models wrapping EfficientNet, we need to access the
    # inner EfficientNet layer. For Functional models, layers are directly
    # accessible by name.

    # Try to get the target layer directly from the model
    try:
        last_conv_layer = model.get_layer(last_conv_layer_name)
    except ValueError:
        # The layer might be inside a sub-model
        last_conv_layer = None
        for layer in model.layers:
            if hasattr(layer, 'get_layer'):
                try:
                    last_conv_layer = layer.get_layer(last_conv_layer_name)
                    print(f"Found '{last_conv_layer_name}' inside sub-model '{layer.name}'")
                    break
                except ValueError:
                    continue

        if last_conv_layer is None:
            raise ValueError(f"Could not find layer '{last_conv_layer_name}' in model.")

    # ── Step 2: Build a Grad-CAM model ────────────────────────────────────────
    # This sub-model takes the original input and returns:
    #   - the output of the last conv layer (feature maps)
    #   - the final classification predictions
    grad_model = keras.Model(
        inputs=model.inputs,
        outputs=[
            last_conv_layer.output,   # Feature maps from last conv layer
            model.output              # Final softmax probabilities
        ]
    )

    # ── Step 3: Forward pass + record gradients ───────────────────────────────
    # tf.GradientTape records all operations so we can compute gradients later
    with tf.GradientTape() as tape:
        # Cast image to float32 (required for gradient computation)
        img_tensor = tf.cast(img_array, tf.float32)

        # Run the image through the model, getting feature maps + predictions
        conv_outputs, predictions = grad_model(img_tensor)

        # If no class specified, use the predicted (highest probability) class
        if pred_index is None:
            pred_index = tf.argmax(predictions[0]).numpy()

        # Get the score for our target class
        # This is the value we will differentiate w.r.t. the conv layer output
        class_channel = predictions[:, pred_index]

    # ── Step 4: Compute gradients ─────────────────────────────────────────────
    # How much does each feature map activation affect the class score?
    # grads shape: (1, H', W', C) where C = number of feature maps
    grads = tape.gradient(class_channel, conv_outputs)

    # ── Step 5: Global Average Pooling of gradients ───────────────────────────
    # Average across the spatial dimensions (H', W') for each feature map
    # pooled_grads shape: (C,) — one importance weight per feature map
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # ── Step 6: Weighted sum of feature maps ──────────────────────────────────
    # Multiply each feature map by its importance weight
    # conv_outputs shape: (1, H', W', C)
    conv_outputs = conv_outputs[0]                      # Remove batch dim → (H', W', C)
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]  # Weighted sum → (H', W', 1)
    heatmap = tf.squeeze(heatmap)                       # Remove last dim → (H', W')

    # ── Step 7: Apply ReLU ────────────────────────────────────────────────────
    # ReLU keeps only POSITIVE activations — regions that increase class score
    # Negative regions push AGAINST this class and are not useful here
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-8)

    # ── Step 8: Return numpy arrays ───────────────────────────────────────────
    heatmap = heatmap.numpy()
    pred_probs = predictions[0].numpy()

    return heatmap, int(pred_index), pred_probs


print("make_gradcam_heatmap() function defined ✅")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 7 — Overlay Heatmap on Original Image
# ═══════════════════════════════════════════════════════════════════════════════

def overlay_heatmap_on_image(original_img, heatmap,
                             colormap=cv2.COLORMAP_JET,
                             alpha=0.45):
    """
    Resizes the Grad-CAM heatmap to match the original image and blends them.

    Process:
    1. Resize heatmap to (H, W) of the original image.
    2. Convert heatmap [0,1] → [0,255] and apply a color map (e.g., JET).
    3. Blend original image and colored heatmap using weighted sum.

    Parameters:
        original_img (np.ndarray): Original RGB image, shape (H, W, 3), uint8.
        heatmap (np.ndarray)     : 2D heatmap from make_gradcam_heatmap(),
                                   values in [0, 1], shape (H', W').
        colormap (int)           : OpenCV colormap constant. Default: COLORMAP_JET
                                   (blue=low, red=high importance).
        alpha (float)            : Heatmap transparency. 0 = invisible, 1 = opaque.
                                   Default 0.45 gives a nice balance.

    Returns:
        superimposed_img (np.ndarray): RGB image with heatmap overlay, uint8.
        heatmap_colored  (np.ndarray): The colored heatmap alone (RGB), uint8.
    """
    H, W = original_img.shape[:2]

    # Step 1: Resize heatmap to match original image size
    heatmap_resized = cv2.resize(heatmap, (W, H))  # Note: cv2 uses (W, H) order

    # Step 2: Scale heatmap to [0, 255] as uint8
    heatmap_uint8 = np.uint8(255 * heatmap_resized)

    # Step 3: Apply color map (JET: blue → green → yellow → red)
    heatmap_colored_bgr = cv2.applyColorMap(heatmap_uint8, colormap)

    # Step 4: Convert heatmap from BGR to RGB (for matplotlib display)
    heatmap_colored = cv2.cvtColor(heatmap_colored_bgr, cv2.COLOR_BGR2RGB)

    # Step 5: Blend original image (as float) with colored heatmap
    # Formula: superimposed = (1-alpha) * original + alpha * heatmap
    original_float = original_img.astype(np.float32)
    heatmap_float  = heatmap_colored.astype(np.float32)

    superimposed = (1 - alpha) * original_float + alpha * heatmap_float

    # Step 6: Clip values to valid range and convert back to uint8
    superimposed_img = np.clip(superimposed, 0, 255).astype(np.uint8)

    return superimposed_img, heatmap_colored


print("overlay_heatmap_on_image() function defined ✅")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 8 — Display Results Function
# ═══════════════════════════════════════════════════════════════════════════════

def display_gradcam_results(original_img, heatmap_colored, superimposed_img,
                            pred_index, pred_probs, class_names,
                            true_label=None, image_path=None):
    """
    Displays a 3-panel visualization:
      [1] Original image
      [2] Grad-CAM heatmap (colored)
      [3] Superimposed image (original + heatmap)

    Also prints a probability bar chart as a 4th panel.

    Parameters:
        original_img     : Original RGB image (H, W, 3).
        heatmap_colored  : Colored heatmap (H, W, 3).
        superimposed_img : Blended image (H, W, 3).
        pred_index (int) : Predicted class index.
        pred_probs       : Softmax probability array for all classes.
        class_names (list): List of class name strings.
        true_label (int) : Ground truth class index (optional).
        image_path (str) : Image filename for display title (optional).
    """
    pred_class = class_names[pred_index]
    confidence = pred_probs[pred_index] * 100

    fig, axes = plt.subplots(1, 4, figsize=(20, 5))
    fig.patch.set_facecolor('#0f0f0f')  # Dark background

    # ── Title bar ─────────────────────────────────────────────────────────────
    img_name = os.path.basename(image_path) if image_path else "Image"
    title_color = '#00ff9f'  # Neon green
    is_correct = ""
    if true_label is not None:
        is_correct = " ✅" if true_label == pred_index else f" ❌ (True: {class_names[true_label]})"

    fig.suptitle(
        f"Grad-CAM Analysis  |  {img_name}\n"
        f"Prediction: {pred_class}  ({confidence:.1f}% confidence){is_correct}",
        color=title_color, fontsize=13, fontweight='bold', y=1.02
    )

    # ── Panel 1: Original Image ───────────────────────────────────────────────
    axes[0].imshow(original_img)
    axes[0].set_title('Original Image', color='white', fontsize=11, pad=8)
    axes[0].axis('off')
    for spine in axes[0].spines.values():
        spine.set_edgecolor('#333')

    # ── Panel 2: Grad-CAM Heatmap ─────────────────────────────────────────────
    axes[1].imshow(heatmap_colored)
    axes[1].set_title('Grad-CAM Heatmap\n(Red = Most Important)', color='white', fontsize=11, pad=8)
    axes[1].axis('off')

    # Add a colorbar legend
    sm = plt.cm.ScalarMappable(cmap='jet', norm=plt.Normalize(vmin=0, vmax=1))
    cbar = plt.colorbar(sm, ax=axes[1], fraction=0.046, pad=0.04)
    cbar.set_label('Activation\nIntensity', color='white', fontsize=8)
    cbar.ax.yaxis.set_tick_params(color='white')
    plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='white', fontsize=7)

    # ── Panel 3: Superimposed ─────────────────────────────────────────────────
    axes[2].imshow(superimposed_img)
    axes[2].set_title('Heatmap Overlay\n(Original + Grad-CAM)', color='white', fontsize=11, pad=8)
    axes[2].axis('off')

    # ── Panel 4: Probability Bar Chart ────────────────────────────────────────
    bar_colors = ['#e74c3c' if i == pred_index else '#3498db'
                  for i in range(len(class_names))]
    bars = axes[3].barh(class_names, pred_probs * 100,
                        color=bar_colors, edgecolor='#555', linewidth=0.5)

    # Add percentage labels on bars
    for bar, prob in zip(bars, pred_probs):
        axes[3].text(
            bar.get_width() + 0.5,
            bar.get_y() + bar.get_height() / 2,
            f'{prob*100:.1f}%',
            va='center', ha='left',
            color='white', fontsize=8
        )

    axes[3].set_xlim(0, 115)
    axes[3].set_xlabel('Confidence (%)', color='#aaa', fontsize=9)
    axes[3].set_title('Class Probabilities', color='white', fontsize=11, pad=8)
    axes[3].tick_params(colors='white', labelsize=9)
    axes[3].set_facecolor('#1a1a1a')
    axes[3].spines['bottom'].set_color('#444')
    axes[3].spines['top'].set_visible(False)
    axes[3].spines['right'].set_visible(False)
    axes[3].spines['left'].set_color('#444')

    # Legend for bar colors
    red_patch  = mpatches.Patch(color='#e74c3c', label='Predicted')
    blue_patch = mpatches.Patch(color='#3498db', label='Others')
    axes[3].legend(handles=[red_patch, blue_patch], loc='lower right',
                   facecolor='#222', edgecolor='#555', labelcolor='white', fontsize=8)

    # Style all axes backgrounds dark
    for ax in axes[:3]:
        ax.set_facecolor('#0f0f0f')

    plt.tight_layout()
    plt.savefig('gradcam_result.png', dpi=150, bbox_inches='tight',
                facecolor='#0f0f0f')
    plt.show()
    print(f"💾 Saved as 'gradcam_result.png'")


print("display_gradcam_results() function defined ✅")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 9 — Master Pipeline Function
# ═══════════════════════════════════════════════════════════════════════════════

def run_gradcam(image_path, model, layer_name, class_names,
                true_label=None, alpha=0.45, colormap=cv2.COLORMAP_JET):
    """
    Master function that runs the full Grad-CAM pipeline for one image.

    Calls all other functions in order:
      1. preprocess_image()
      2. make_gradcam_heatmap()
      3. overlay_heatmap_on_image()
      4. display_gradcam_results()

    Parameters:
        image_path  (str)  : Path to the input image.
        model              : Trained Keras model.
        layer_name  (str)  : Name of the Grad-CAM target layer.
        class_names (list) : List of class name strings.
        true_label  (int)  : Ground truth class index (optional, for evaluation).
        alpha       (float): Overlay transparency (0.0–1.0).
        colormap           : OpenCV colormap for heatmap coloring.

    Returns:
        dict: A dictionary containing all results for further analysis.
    """
    print(f"\n{'='*60}")
    print(f" Processing: {os.path.basename(image_path)}")
    print(f"{'='*60}")

    # ── Step 1: Preprocess ────────────────────────────────────────────────────
    img_array, original_img = preprocess_image(image_path, img_size=IMG_SIZE)
    print(f" ✅ Image loaded and preprocessed → shape: {img_array.shape}")

    # ── Step 2: Generate Grad-CAM heatmap ─────────────────────────────────────
    heatmap, pred_index, pred_probs = make_gradcam_heatmap(
        img_array, model, layer_name
    )
    print(f" ✅ Grad-CAM heatmap generated → heatmap shape: {heatmap.shape}")
    print(f" 🎯 Predicted class: {class_names[pred_index]} "
          f"({pred_probs[pred_index]*100:.1f}% confidence)")

    # ── Step 3: Overlay heatmap on image ──────────────────────────────────────
    superimposed_img, heatmap_colored = overlay_heatmap_on_image(
        original_img, heatmap, colormap=colormap, alpha=alpha
    )
    print(f" ✅ Heatmap overlaid on image")

    # ── Step 4: Display results ───────────────────────────────────────────────
    display_gradcam_results(
        original_img, heatmap_colored, superimposed_img,
        pred_index, pred_probs, class_names,
        true_label=true_label, image_path=image_path
    )

    # Return all results as a dict for further use
    return {
        'image_path'      : image_path,
        'original_img'    : original_img,
        'heatmap'         : heatmap,
        'heatmap_colored' : heatmap_colored,
        'superimposed'    : superimposed_img,
        'pred_index'      : pred_index,
        'pred_class'      : class_names[pred_index],
        'pred_probs'      : pred_probs,
        'confidence'      : pred_probs[pred_index],
        'true_label'      : true_label,
    }


print("run_gradcam() master function defined ✅")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 10 — Run Grad-CAM on a SINGLE Image
# ═══════════════════════════════════════════════════════════════════════════════

# ── Change this to the path of any retinal image you want to test ─────────────
TEST_IMAGE_PATH = 'F:/DR_Project/data/train_images/Moderate/0024cdab0c1e.png'

# Optional: provide the true label if you know it (for evaluation)
# Use None if you don't want to show ground truth
TRUE_LABEL = 1   # 0=Mild, 1=Moderate, 2=No_DR, 3=Proliferate_DR, 4=Severe

# Run the full Grad-CAM pipeline
result = run_gradcam(
    image_path=TEST_IMAGE_PATH,
    model=model,
    layer_name=gradcam_layer_name,
    class_names=CLASS_NAMES,
    true_label=TRUE_LABEL,
    alpha=0.45           # Heatmap transparency (0=invisible, 1=full)
)

# You can access individual results:
print(f"\n📊 Summary:")
print(f"  Predicted class : {result['pred_class']}")
print(f"  Confidence      : {result['confidence']*100:.2f}%")
print(f"  All probabilities:")
for name, prob in zip(CLASS_NAMES, result['pred_probs']):
    bar = '█' * int(prob * 30)
    print(f"    {name:<18} {bar:<30} {prob*100:5.1f}%")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 11 — Run Grad-CAM on MULTIPLE Test Images
# ═══════════════════════════════════════════════════════════════════════════════

def run_gradcam_batch(image_paths, model, layer_name, class_names,
                      true_labels=None, save_dir='gradcam_outputs'):
    """
    Runs Grad-CAM on a list of images and saves all results.

    Parameters:
        image_paths (list) : List of image file paths.
        model              : Trained Keras model.
        layer_name  (str)  : Target layer for Grad-CAM.
        class_names (list) : List of class name strings.
        true_labels (list) : Optional list of true label indices.
        save_dir    (str)  : Directory to save output images.

    Returns:
        list: List of result dicts for all images.
    """
    # Create output directory
    os.makedirs(save_dir, exist_ok=True)
    print(f"📁 Output directory: '{save_dir}'")

    all_results = []

    for i, img_path in enumerate(image_paths):
        true_label = true_labels[i] if true_labels else None

        try:
            # Preprocess
            img_array, original_img = preprocess_image(img_path)

            # Grad-CAM
            heatmap, pred_index, pred_probs = make_gradcam_heatmap(
                img_array, model, layer_name
            )

            # Overlay
            superimposed_img, heatmap_colored = overlay_heatmap_on_image(
                original_img, heatmap
            )

            # ── Save side-by-side (no plt.show() to avoid popup for each image)
            fig, axes = plt.subplots(1, 3, figsize=(15, 5))
            axes[0].imshow(original_img)
            axes[0].set_title('Original', fontsize=10)
            axes[0].axis('off')
            axes[1].imshow(heatmap_colored)
            axes[1].set_title('Grad-CAM', fontsize=10)
            axes[1].axis('off')
            axes[2].imshow(superimposed_img)
            axes[2].set_title('Overlay', fontsize=10)
            axes[2].axis('off')

            pred_class = class_names[pred_index]
            confidence = pred_probs[pred_index] * 100
            correct_str = ""
            if true_label is not None:
                correct_str = " ✅" if true_label == pred_index else f" ❌(True:{class_names[true_label]})"

            fig.suptitle(f"Pred: {pred_class} ({confidence:.1f}%){correct_str}",
                         fontsize=12)
            plt.tight_layout()

            # Save to file
            base_name = os.path.splitext(os.path.basename(img_path))[0]
            save_path = os.path.join(save_dir, f"{base_name}_gradcam.png")
            plt.savefig(save_path, dpi=120, bbox_inches='tight')
            plt.close()

            all_results.append({
                'image_path' : img_path,
                'pred_class' : pred_class,
                'confidence' : confidence,
                'true_label' : true_label,
                'correct'    : (true_label == pred_index) if true_label is not None else None,
                'saved_to'   : save_path
            })

            print(f"[{i+1:3d}/{len(image_paths)}] {os.path.basename(img_path):<30} "
                  f"→ {pred_class:<18} ({confidence:.1f}%){correct_str}")

        except Exception as e:
            print(f"[{i+1:3d}/{len(image_paths)}] ❌ Error on {img_path}: {e}")
            all_results.append({'image_path': img_path, 'error': str(e)})

    # Summary
    print(f"\n{'='*60}")
    print(f" ✅ Batch complete. Processed {len(image_paths)} images.")
    if true_labels:
        n_correct = sum(r.get('correct', False) for r in all_results if 'correct' in r)
        print(f" 🎯 Accuracy: {n_correct}/{len(image_paths)} correct")
    print(f" 💾 Results saved to: '{save_dir}/'")
    print(f"{'='*60}")

    return all_results


# ── Example: Run on multiple test images ─────────────────────────────────────
# Uncomment and modify these paths to test multiple images at once:

# test_images = [
#     'F:/DR_Project/data/train_images/Mild/001639a390f0.png',
#     'F:/DR_Project/data/train_images/Moderate/0024cdab0c1e.png',
#     'F:/DR_Project/data/train_images/No_DR/002c21358ce6.png',
#     'F:/DR_Project/data/train_images/Severe/004e48aec2c0.png',
#     'F:/DR_Project/data/train_images/Proliferate_DR/005b95c28852.png',
# ]
# true_labels = [0, 1, 2, 4, 3]  # Mild=0, Moderate=1, No_DR=2, Prolif=3, Severe=4

# batch_results = run_gradcam_batch(
#     image_paths=test_images,
#     model=model,
#     layer_name=gradcam_layer_name,
#     class_names=CLASS_NAMES,
#     true_labels=true_labels,
#     save_dir='gradcam_outputs'
# )

print("run_gradcam_batch() function defined ✅")
print("Uncomment the example above to test multiple images.")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 12 — Auto-Sample: Show One Grad-CAM per Class
# ═══════════════════════════════════════════════════════════════════════════════

def show_gradcam_per_class(base_dir, model, layer_name, class_names,
                            n_per_class=1, imgs_per_class_extension='.png'):
    """
    Automatically picks one image from each class folder and
    generates a Grad-CAM visualization for it.

    Useful for a quick overview of how the model handles each severity.

    Parameters:
        base_dir    (str)  : Root directory containing class subfolders.
        model              : Trained Keras model.
        layer_name  (str)  : Target conv layer name.
        class_names (list) : Class names (must match subfolder names).
        n_per_class (int)  : Number of images to sample per class.
    """
    print("🔍 Auto-sampling one image per class...\n")

    for class_idx, class_name in enumerate(class_names):
        class_dir = os.path.join(base_dir, class_name)

        if not os.path.exists(class_dir):
            print(f"⚠️  Folder not found: {class_dir}")
            continue

        # Get image files
        files = [f for f in os.listdir(class_dir)
                 if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        if not files:
            print(f"⚠️  No images found in: {class_dir}")
            continue

        # Pick the first n_per_class images
        for i in range(min(n_per_class, len(files))):
            img_path = os.path.join(class_dir, files[i])
            run_gradcam(
                image_path=img_path,
                model=model,
                layer_name=layer_name,
                class_names=class_names,
                true_label=class_idx
            )


# ── Uncomment to auto-show one image per DR severity class ─────────────────
# show_gradcam_per_class(
#     base_dir=TEST_IMG_DIR,
#     model=model,
#     layer_name=gradcam_layer_name,
#     class_names=CLASS_NAMES,
#     n_per_class=1
# )

print("show_gradcam_per_class() function defined ✅")
print("Uncomment the call above to auto-generate one Grad-CAM per class.")
```

```python
# ═══════════════════════════════════════════════════════════════════════════════
# CELL 13 — (Bonus) Compare Grad-CAM for Different Target Classes
#            on the SAME image
# ═══════════════════════════════════════════════════════════════════════════════

def compare_gradcam_all_classes(image_path, model, layer_name, class_names):
    """
    Generates Grad-CAM for ALL classes on the same image.
    This shows which regions the model focuses on for EACH class —
    great for viva demonstration.

    Parameters:
        image_path  (str) : Path to the input image.
        model             : Trained Keras model.
        layer_name  (str) : Target conv layer name.
        class_names (list): List of class names.
    """
    # Preprocess once
    img_array, original_img = preprocess_image(image_path)

    n_classes = len(class_names)
    fig, axes = plt.subplots(2, n_classes, figsize=(4 * n_classes, 9))
    fig.patch.set_facecolor('#111')

    fig.suptitle(
        f"Grad-CAM for All Classes  |  {os.path.basename(image_path)}\n"
        f"Top row: Heatmap  |  Bottom row: Overlay",
        color='white', fontsize=13, fontweight='bold'
    )

    for class_idx in range(n_classes):
        # Generate heatmap for this specific class
        heatmap, pred_idx, pred_probs = make_gradcam_heatmap(
            img_array, model, layer_name, pred_index=class_idx
        )
        superimposed, heatmap_colored = overlay_heatmap_on_image(
            original_img, heatmap
        )

        confidence = pred_probs[class_idx] * 100
        is_predicted = "⭐ " if class_idx == pred_idx else ""

        # Row 0: Heatmap
        axes[0, class_idx].imshow(heatmap_colored)
        axes[0, class_idx].set_title(
            f"{is_predicted}{class_names[class_idx]}\n{confidence:.1f}%",
            color='#00ff9f' if class_idx == pred_idx else 'white',
            fontsize=9
        )
        axes[0, class_idx].axis('off')

        # Row 1: Overlay
        axes[1, class_idx].imshow(superimposed)
        axes[1, class_idx].axis('off')

    plt.tight_layout()
    save_name = f"gradcam_all_classes_{os.path.splitext(os.path.basename(image_path))[0]}.png"
    plt.savefig(save_name, dpi=130, bbox_inches='tight', facecolor='#111')
    plt.show()
    print(f"💾 Saved as '{save_name}'")


# ── Uncomment to compare all classes on one image ──────────────────────────
# compare_gradcam_all_classes(
#     image_path=TEST_IMAGE_PATH,
#     model=model,
#     layer_name=gradcam_layer_name,
#     class_names=CLASS_NAMES
# )

print("compare_gradcam_all_classes() function defined ✅")
print("Uncomment the call above to compare Grad-CAM across all 5 severity classes.")
```

## 🎓 Viva Tips

**Q: What does red/yellow mean in Grad-CAM?**  
A: Red = highest importance (most influential to the prediction), blue = least important. The JET colormap goes blue → green → yellow → red.

**Q: Why use the last convolutional layer (`top_conv`)?**  
A: The last conv layer has the richest semantic features — it "understands" high-level concepts like lesions. Earlier layers only detect edges/textures which are less class-specific.

**Q: Why apply ReLU to the heatmap?**  
A: We only want regions that *increase* the class score. Negative values suppress the class and are not relevant for explaining the positive prediction.

**Q: How is Grad-CAM different from saliency maps?**  
A: Saliency maps compute gradients w.r.t. the *input pixels* (noisy, hard to interpret). Grad-CAM uses gradients w.r.t. *feature maps* in a specific layer, producing smoother, more meaningful localization maps.

**Q: Does Grad-CAM change the model?**  
A: No! Grad-CAM is a *post-hoc* explanation method — it reads from an already-trained model without modifying it.
